```python
from qwen_rag_eval.runner import DefaultRunner
from qwen_rag_eval import load_cmrc_samples
from qwen_rag_eval.evaluation import RagBatchRunner, RagasEvaluator

if __name__ == "__main__":
    # 1. 加载评估样本（假设你已经用 build_eval_samples 生成了 samples）
    samples = load_cmrc_samples("config/application.yaml")

    # 2. 默认 RAG runner
    runner = DefaultRunner()

    # 3. 批量执行 RAG
    batch = RagBatchRunner(runner, mode="default")
    records = batch.run_batch(samples, limit=200, show_progress=True)

    # 4. RAGAS 评估
    evaluator = RagasEvaluator("config/application.yaml")
    result = evaluator.evaluate(records)

    # 5. 查看结果
    result.show_console()
```
